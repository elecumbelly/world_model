# Explore Grid World

Quick exploration of the 16x16 grid world environment.

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import matplotlib.pyplot as plt
from world_model.envs.grid_world import GridWorld
```

## Create and Reset Environment

```python
env = GridWorld(grid_size=16, render_size=64, max_steps=200)
obs = env.reset(seed=42)

print(f"Observation shape: {obs.shape}")
print(f"Action space: {env.action_space_n} actions")
print(f"Agent position: {env.agent_pos}")

plt.figure(figsize=(4, 4))
plt.imshow(obs)
plt.title('Initial Grid World')
plt.axis('off')
plt.show()
```

## Multiple Seeds

```python
fig, axes = plt.subplots(2, 4, figsize=(12, 6))
for i, ax in enumerate(axes.flat):
    obs = env.reset(seed=i * 10)
    ax.imshow(obs)
    ax.set_title(f'Seed {i * 10}')
    ax.axis('off')
plt.suptitle('Different Map Seeds')
plt.tight_layout()
plt.show()
```

## Random Rollout

```python
obs = env.reset(seed=42)
observations = [obs]
rewards = []
rng = np.random.RandomState(42)

for _ in range(50):
    action = rng.randint(0, 4)
    obs, reward, done, info = env.step(action)
    observations.append(obs)
    rewards.append(reward)
    if done:
        break

print(f"Rollout length: {len(rewards)} steps")
print(f"Total reward: {sum(rewards):.2f}")
print(f"Cumulative rewards: {np.cumsum(rewards)[-1]:.2f}")

# Show trajectory frames
fig, axes = plt.subplots(1, 6, figsize=(15, 3))
for i, ax in enumerate(axes):
    idx = i * (len(observations) // 6)
    ax.imshow(observations[idx])
    ax.set_title(f't={idx}')
    ax.axis('off')
plt.suptitle('Random Rollout Trajectory')
plt.tight_layout()
plt.show()
```

## Reward Distribution (Random Policy)

```python
episode_rewards = []
episode_lengths = []

for seed in range(100):
    obs = env.reset(seed=seed)
    total = 0
    steps = 0
    done = False
    rng = np.random.RandomState(seed)
    while not done:
        obs, reward, done, info = env.step(rng.randint(0, 4))
        total += reward
        steps += 1
    episode_rewards.append(total)
    episode_lengths.append(steps)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.hist(episode_rewards, bins=30)
ax1.set_title(f'Episode Rewards (mean={np.mean(episode_rewards):.2f})')
ax1.set_xlabel('Total Reward')
ax2.hist(episode_lengths, bins=30)
ax2.set_title(f'Episode Lengths (mean={np.mean(episode_lengths):.0f})')
ax2.set_xlabel('Steps')
plt.tight_layout()
plt.show()
```
